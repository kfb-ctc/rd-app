```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('project_ledger.xlsx', sheet_name = 'ERCIP Obligation Ledger', header = 3, usecols=lambda x: 'Unnamed' not in x)
df.head()
```

```python
project_list = df['Project #'].unique()
```

```python
df_funding = pd.DataFrame(columns = ['Project','Original_funding','Funding_actions'])
for project in project_list:
    project_df = df[df['Project #'] == project]
    project_original_funding = project_df['P&D Original Programmed Amount'].sum(axis = 1, skipna = True) + project_df['Construction Original Programmed Amount'].sum(axis = 1, skipna = True)
    project_funding_actions = project_df['P&D Funding Actions'].sum(axis = 1, skipna = True) + project_df['Construction Funding Actions'].sum(axis = 1, skipna = True)
    row = {'Project': project_df['Project #'],'Original_funding': project_original_funding,'Funding_actions': project_funding_actions}
    df_funding.append(row)
df_funding.head()
```

```python
project_df = df[df['Project #'] == project_list[0]]
project_df
```

```python
project_original_funding = project_df['P&D Original Programmed Amount'].sum() + project_df['Construction Original Programmed Amount '].sum()
project_funding_actions = project_df['P&D Funding Actions'].sum() + project_df['Construction Funding Actions'].sum()
print(project_original_funding)
print(project_funding_actions)
```
